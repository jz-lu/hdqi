```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True

# Create x values from 0 to 1
x = np.linspace(0, 1, 400)

# --- Original curves for different block sizes ---
k_small = 40
y_small = 1 / (1 + np.exp(k_small * (x - 0.1)))

k_large = 150
y_large = 1 / (1 + np.exp(k_large * (x - 0.1)))

# --- Define the asymptotic line as a true step function ---
# y is 1.0 where x < 0.1, and 0.0 otherwise.
y_step = np.where(x < 0.1, 1.0, 0.0)

# Create the plot
plt.figure(figsize=(8, 6))
plt.rc("font", size=16)

# Plot the functions
plt.plot(x, y_small, 'b-', lw=2, label='Small Block Size')
plt.plot(x, y_large, 'g-', lw=2, label='Large Block Size')
plt.plot(x, y_step, 'r--', lw=1, label='Asymptotic')

# Customize the plot
plt.xlabel('Fraction of bits flipped', fontsize=16)
plt.ylabel('Decoder success rate', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=14)
plt.xlim(0, 0.5)
plt.ylim(-0.05, 1.05)

# Save the figure
plt.savefig('waterfall.pdf')
```
